**<h2>Data Cleaning</h2>**

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df
```

```python
df.loc[:, 'days'] = pd.to_datetime(df['date']).dt.day_name()
df
```

```python
df['year'] = df['date'].dt.year
df
```

```python
df = df.loc[(df['date'].dt.year >= 2020) & (df['date'].dt.year <= 2024)]
df
```

```python
df = df.loc[:, ~df.columns.isin(['type_apheresis_platelet', 'type_apheresis_plasma',
                                  'type_other', 'social_civilian', 'social_student',
                                  'social_policearmy', 'state'])]
df
```

```python
# df.loc[:, ['daily', 'blood_a', 'blood_b', 'blood_o', 'blood_ab',
#            'donations_new', 'donations_regular', 'donations_irregular', 'year']] = \
# df.loc[:, ['daily', 'blood_a', 'blood_b', 'blood_o', 'blood_ab',
#            'donations_new', 'donations_regular', 'donations_irregular', 'year']].astype('int16')
# df.info()
df[['daily', 'blood_a', 'blood_b', 'blood_o', 'blood_ab',
           'donations_new', 'donations_regular', 'donations_irregular', 'year']] = df[['daily', 'blood_a', 'blood_b', 'blood_o', 'blood_ab',
           'donations_new', 'donations_regular', 'donations_irregular', 'year']].astype('int16')
df.info()
```

**<h2>Data Visualization</h2>**

```python
# df = df.copy()
blood_types = ['blood_a', 'blood_b', 'blood_o', 'blood_ab']

years = df['year'].unique()
x = np.arange(len(blood_types))

plt.figure(figsize=(14, 9))

colors = ['blue', 'red', 'green', 'purple', 'orange', 'brown']

for i, year in enumerate(sorted(years)):
    df_year = df[df['year'] == year]
    avg_blood_donations = [df_year[blood].sum() for blood in blood_types]

    plt.bar(x + (i * 0.15), avg_blood_donations, width=0.15, label=f'Year {year}', color=colors[i % len(colors)])

plt.xlabel("Blood Type")
plt.ylabel("Total Donations")
plt.title("Blood Donations by Blood Type for Each Year")
plt.xticks(x + 0.15, blood_types)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
blood_types = ['daily', 'blood_a', 'blood_b', 'blood_o', 'blood_ab']
df_blood = df[blood_types]

plt.figure(figsize=(14, 9))
sns.boxplot(data=df_blood)
plt.title("Boxplot of Blood Donations (Before Outlier Removal)")
plt.show()
```

```python
blood_columns = ['daily', 'blood_a', 'blood_b', 'blood_o', 'blood_ab']
df_blood = df[blood_columns]

Q1 = df_blood.quantile(0.25)
Q3 = df_blood.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_blood_filtered = df_blood[~((df_blood < lower_bound) | (df_blood > upper_bound)).any(axis=1)]
```

```python
df
```

```python
df_final = df.loc[df_blood_filtered.index]
df_final
```

```python
df_final.describe()
```

```python
plt.figure(figsize=(14, 9))
sns.boxplot(data=df_blood_filtered)
plt.title("Boxplot of Blood Donations (After Outlier Removal)")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_final['daily'], kde=True, bins=30)
plt.axvline(df_final['daily'].mean(), color='red', linestyle='dashed', label="Mean")
plt.axvline(df_final['daily'].median(), color='blue', linestyle='dashed', label="Median")
plt.legend()
plt.title("Distribution of Daily Donations")
plt.show()
# Why? Helps to visualize the distribution of blood donations (daily, new, regular, irregular) and detect skewness.
```

```python
blood_types = ["blood_a", "blood_b", "blood_o", "blood_ab"]

plt.figure(figsize=(12, 8))

for i, blood in enumerate(blood_types, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=df_final["daily"], y=df_final[blood])
    plt.title(f"Daily Donations vs {blood.replace('_', ' ').title()}")
    plt.xlabel("Daily Donations")
    plt.ylabel(f"{blood.replace('_', ' ').title()} Donations")

plt.tight_layout()
plt.show()
# Why? To check the relationship between daily donations and a specific blood type.
# Insight: Identifies trends and potential patterns.
```

```python
blood_sums = df_final[['blood_a', 'blood_b', 'blood_o', 'blood_ab']].sum()
plt.figure(figsize=(8, 8))
plt.pie(blood_sums, labels=blood_sums.index, autopct="%1.1f%%", colors=['red', 'blue', 'green', 'purple'])
plt.title("Proportion of Blood Donations by Type")
plt.show()
# Why? If you just want to compare the proportion of donations by blood type instead of trends.
# Best for: Showing the percentage of each blood type donated.
# Insight: Easily shows which blood type is donated the most and least.
```

```python
# Exclude the "days" column before computing correlation
df_numeric = df_final.select_dtypes(include=["number"])

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Blood Donation Features")
plt.show()
```
